```python
#读数程序
import numpy as np
def dushu(lieshu):
    fp=open("C:/SNLO/PWMIX_LP.DAT",'r')
    data=[]
    ls=[]
    a=[]
    c=[]
    d=lieshu #y是第几列数据
    num=0
    for line in fp:
        line = line.replace("\n","")
        line = line.replace("    "," ")
        line = line.replace("   "," ")
        line = line.replace("  "," ")
            #print(line)
        a1=line.split(' ')
        #print(a1)
        b=[]
        for x in a1:
            if x.strip()!='':
                b.append(x.strip())   
        c.append(b) 
        num=num+1 #一共有多少列数据
        #将空格作为分隔符将一个字符切割成一个字符数组
    fp.close()
    #ls=numpy.array(ls,dtype=float)   #将其转换成numpy的数组，并定义数据类型为float
    for i in range(0,num):
        a.append(eval(c[i][d]))
    x1=np.array(a, dtype=np.float64)
    return x1
xx=dushu(0)
y1=dushu(3)
print(xx)
print(y1)
```
