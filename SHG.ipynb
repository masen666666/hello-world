```python
import  numpy as np
import matplotlib.pyplot as plt
 
x = np.linspace(-10, 10, 1000)
a = np.sinc(x)**2

# d = np.log(x)
 
plt.figure(figsize=(8,4))
plt.plot(x,a,label='$sinc(deltak)**2$',color='green',linewidth=0.5)

# plt.plot(x,d,label='$log(x)$',color='grey',linewidth=0.5)
 
plt.xlabel('deltak')
plt.ylabel('I2')
plt.xlim(-5,5)
plt.ylim(-0.1,1.2)
plt.legend()
plt.show()
```
