```python
#image
from astropy.io import fits
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt,exp
hdu = fits.open('E:/python lianxi/lianxi5/mystery.fits')
img=hdu[0].data
def distance(point1,point2):
    return sqrt((point1[0]-point2[0])**2 + (point1[1]-point2[1])**2)

def idealFilterLP(D0,imgShape):
    base = np.zeros(imgShape[:2])
    rows, cols = imgShape[:2]
    center = (rows/2,cols/2)
    for x in range(cols):
        for y in range(rows):
            if distance((y,x),center) < D0:
                base[y,x] = 1
    return base

#np.fft.fft2(img)
#np.fft.ifft2(LowPass)
#np.fft.fftshift(original)
LowPass = idealFilterLP(10,img.shape)
plt.figure(figsize=(6.4*5, 4.8*5), constrained_layout=False)
o1=np.fft.fft2(img)
o2=np.fft.fftshift(o1)
tu1=np.fft.fft2(LowPass)
tu2=np.fft.fftshift(tu1)
i1=np.fft.fftshift(np.fft.ifft2(tu2*o2))

plt.subplot(152), plt.imshow(np.log(np.abs(o2)), "gray"), plt.title("fast fourier transform object")
plt.subplot(155), plt.imshow(np.abs(i1), "gray"), plt.title("image")
plt.subplot(151), plt.imshow(img, "gray"), plt.title("Object")
plt.subplot(153), plt.imshow(np.abs(LowPass), "gray"), plt.title("aperture")
plt.subplot(154), plt.imshow(np.log(np.abs(tu2)), "gray"), plt.title("far field")

```

```python
import numpy as np
import matplotlib.pyplot as plt
def r(x,y):
    """Just for convenience, a function that returns r=sqrt(x^2+y^2)"""
    return np.sqrt(xx**2+yy**2)
def hole(R):
    x = np.linspace(-532,532,1064)
    y = np.linspace(-532,532,1064)
    xx, yy = np.meshgrid(x,y)
    k=r(xx,yy)
    z = np.piecewise(k,[k<R,k>R],[1,0])
    return z
#np.fft.fft2(img)
#np.fft.ifft2(LowPass)
#np.fft.fftshift(original)
kk=hole(32)
fkk=np.fft.fftshift(np.fft.fft2(kk))
plt.figure(figsize=(6.4*5, 4.8*5), constrained_layout=False)
plt.subplot(121), plt.imshow(kk, "gray"), plt.title("aperture")
plt.subplot(122), plt.imshow(np.log(1+np.abs(fkk)), "gray"), plt.title("PSF")
```
